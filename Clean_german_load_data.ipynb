```python
import pandas as pd
import numpy as np
import matplotlib as mb
import seaborn as sb
```

```python

url = 'https://data.open-power-system-data.org/time_series/2020-10-06/time_series_60min_singleindex.csv'
load_timeseries_60min = pd.read_csv(url, parse_dates=['utc_timestamp'], index_col='utc_timestamp')
#parsing the timestamp, so resampling and other functions can be leveraged
```

```python
load_df = load_timeseries_60min #copy of the data from web
```

```python
load_df.head()
```

```python
load_df = load_df.drop(columns = ['cet_cest_timestamp'])
load_df.head()
```

```python
load_time= load_df.tz_convert(None)
load_time.head()
```

```python
load_time.index.name ='Time'
load_time.head()
```

```python
load_de = load_time.filter(like='DE_').copy()
load_de.head()
```

```python
load_de["Total_renewable_energy"] = load_de['DE_solar_generation_actual'].fillna(0)+load_de['DE_wind_generation_actual'].fillna(0)
load_de.head()
```

```python
load_de = load_de.dropna(subset='DE_load_actual_entsoe_transparency')
load_de.describe()
```

```python
load_de['hour'] = load_de.index.hour
load_de['day'] = load_de.index.day
load_de['weekday'] = load_de.index.weekday
load_de['month'] = load_de.index.month
load_de.head()
```

```python
load_de['lag_1h'] = load_de['DE_load_actual_entsoe_transparency'].shift(1)
load_de['lag_24h'] = load_de['DE_load_actual_entsoe_transparency'].shift(24)
load_de.head()

```

```python
# Drop rows with NaNs caused by shift
load_cleaned = load_de
load_cleaned = load_cleaned.dropna(subset=['lag_1h', 'lag_24h'])
load_cleaned.fillna(0)
load_cleaned.head()
```

```python
load_cleaned.to_csv("cleaned_germany_load_data.csv",index=True)
```
